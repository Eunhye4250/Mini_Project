## 여행할 지역 입력

```python
keyword = str(input("여행할 지역을 입력하세요 : "))
search = keyword +' 여행 브이로그 vlog'
url = 'https://www.youtube.com/results?search_query={}'.format(search)
```

#### 화면 불러오기

```python
driver = webdriver.Chrome()
driver.get(url)
soup = bs(driver.page_source, 'html.parser')
#driver.close()
```

#### PAGE_DOWN을 통해 스크롤 내리기 
- body 태그를 찾아 페이지 정보 불러오기

```python
body = driver.find_element_by_tag_name('body')
body.send_keys(Keys.PAGE_DOWN)
```

```python
for i in range(1, 201) :
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
```

```python
soup = bs(driver.page_source, 'lxml')
```

#### 영상 제목 가져오기

```python
title = soup.select('a#video-title')
video_url = soup.select('a#video-title')

# for i in title : 
#     print((i.text.strip()))
```

```python
title_list = []
url_list = []
#jeju_url = []

for i in range(len(title)) :
    title_list.append(title[i].text.strip())
    #url_list.append(i.get('herf'))
    #url_list.append('{}{}'.format('https://www.youtube.com/',i.get('herf')))
    
for i in video_url :
    url_list.append('{}{}'.format('https://www.youtube.com',i.get('href')))
    
print(title_list)
print(url_list)
#print(title_url)

driver.close()
```

```python
info = {
    '영상제목' : title_list,
    '영상주소' : url_list}
pd.DataFrame(info)
```

#### 데이터 프레임으로 변환 후 엑셀로 저장

```python
youtube_df = pd.DataFrame(info)

youtube_df.to_excel(keyword + ".xlsx")
```

```python
title_lst = youtube_df['영상제목'].values.tolist()
url_lst = youtube_df['영상주소'].values.tolist()
```

## DB저장

```python
import psycopg2
```

```python
conn_str = "host=localhost dbname =postgres user=postgres password=admin port = 5432"

try : 
    conn = psycopg2.connect(conn_str)
    print("==== 접속 성공 ====")
except psycopg2.DatabaseError as db_err :
    print("접속오류 !!")
    print(db_err)
```

```python
cur = conn.cursor()
```

```python
for i in range(len(title_lst)) :

    try :
        cur.execute("INSERT INTO trip(title, url_list) VALUES (%s, %s);",
        (title_lst[i], url_lst[i]))
        conn.commit()
        print("----- insert ok")
    except : 
        conn.rollback()
        print("---DB error")
```
